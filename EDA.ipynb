```python
#getting the essential libraries we want:
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#TO GET A HISTOGRAM PLOT FOR A COLUMN
#OR TO CHECK WHETHER RURAL OR URBAN DOMINATES IN THIS CASE

df["Urban_or_Rural"].hist(bins=100)

#Shows median, quartiles, and outliers for the column.
sns.boxplot(x=df["Exceeded_Recommended_Limit"])
#from this we understand Exceeded_Recommended_Limit=True, and a smaller portion are False.

#we do the same for age
sns.boxplot(x=df["Age"])
```

```python
"""If the data type is categorical, then replace it with MODE
else if numerical then with mean
If numerical data has an outlier, go with the median"""
```

```python
median1=df["Age"].median()
median2=df["Exceeded_Recommended_Limit"].median()
median3=df["Educational_to_Recreational_Ratio"].median()


""" A value is trying to be set on a copy of a DataFrame or Series through chained assignment using an inplace method.
The behavior will change in pandas 3.0. This inplace method will never work because the intermediate object on which we are setting values always behaves as a copy."""



df["Agef"] = df["Age"].replace(np.nan,median1)
df["Exceeded_Recommended_Limitf"] = df["Exceeded_Recommended_Limit"].replace(np.nan,median2)
df["Educational_to_Recreational_Ratiof"] = df["Educational_to_Recreational_Ratio"].replace(np.nan,median3)


mode1=df["Gender"].mode()
mode2=df["Primary_Device"].mode()
mode4=df["Urban_or_Rural"].mode()


df["Genderf"] = df["Gender"].replace(np.nan, mode1[0])
df["Primary_Devicef"] = df["Primary_Device"].replace(np.nan, mode2[0])
df["Urban_or_Ruralf"] = df["Urban_or_Rural"].replace(np.nan, mode4[0])

```

```python
sns.countplot(x="Age", hue="Gender", data=df)
plt.title("Age distribution by Gender")
plt.show()
```
